```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('default.payment.next.month',axis=1)
```

```python
X
```

```python
y=df['default.payment.next.month']
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg=LinearRegression()
lreg.fit(X_train,y_train)
```

```python
lreg_y_pred_train=lreg.predict(X_train)
lreg_y_pred_test=lreg.predict(X_test)
```

```python
lreg_y_pred_train
```

```python
from sklearn.metrics import precision_score,accuracy_score,f1_score,r2_score,mean_absolute_error,mean_squared_error
```

```python
accuracy_score(y_train,lreg_y_pred_train)
```

```python
accuracy_score(y_test,lreg_y_pred_test)
```

```python
precision_score(y_train,lreg_y_pred_train)
```

```python
lreg.score(X,y)
```

```python
list(lreg.coef_)
```

```python
lreg.intercept_
```

```python
r2_score(y_test,lreg_y_pred_test)
```

```python
from sklearn.feature_selection import mutual_info_regression
mutual_info=mutual_info_regression(X_train,y_train)
mutual_info
```

```python
mutual_info=pd.Series(mutual_info)
mutual_info.index=X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(15,5))
```

```python
from sklearn.feature_selection import SelectPercentile
```

```python
selected_top_columns=SelectPercentile(mutual_info_regression,percentile=60)
selected_top_columns.fit(X_train,y_train)
```

```python
selected_top_columns.get_support()
```

```python
X_train.columns[selected_top_columns.get_support()]
```
